```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Undirected Erdos-Renyi graphs with n=10

### p=0.05

```python
fig = plt.figure(figsize=(12,12))

for i in range(1,17):
    ax = fig.add_subplot(4,4,i)
    eg = nx.erdos_renyi_graph(n=10,p=0.05)
    nx.draw_networkx(eg)
    
plt.tight_layout()
plt.show()
```
